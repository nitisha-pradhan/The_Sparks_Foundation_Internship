```python
#Importing the required libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
```

```python
import sklearn.datasets as datasets

#Loading the iris dataset from the skleran.datasets library
iris = datasets.load_iris()

#df will have all the features of the dataset.
df = pd.DataFrame(iris.data, columns=iris.feature_names)

#y will contain our class labels that we need to classify.
y=iris.target
```

<h2>Decision Tree Algorithm</h2>
<h3> A decision tree is a flowchart-like tree structure where an internal node represents feature, the branch represents a decision rule, and each leaf node represents the outcome. We split the nodes on the basis of the feature value.<br>
It is a flowchart-like structure helps us in decision making. </h3>

```python
#Splitting the dataset into the train and test sets.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=1)
```

```python
#Create Decision Tree classifer
clf = DecisionTreeClassifier()

#Train Decision Tree Classifer on the train set
clf = clf.fit(X_train, y_train)

#Predict on the test set
y_pred = clf.predict(X_test)
```

```python
#Evaluating the classifier on the test set
#This is data that has not been seen by the classifier. Thus, we can evaluate how accurately the tree can classify new data.
#The accuracy is 96.67% which indicates that the tree is good at predicting the new data.

from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

<h3>Visualizing Decision Tree</h3>

```python
#Importing these for visualizing the tree.

from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```
